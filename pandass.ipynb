6. How many not realized shot on targets each team have

```python
euro["notreal"] = euro["Shots on target"] - euro["Goals"]
euro.groupby('Team').notreal.sum()
```

7. Persent of realizeted shot on targets 

```python
def realization (row):
    row["notreal"] = 1 - (int(row["notreal"]) / int(row["Shots on target"]))
    return row

euro.apply(realization, axis='columns').groupby("Team").notreal.sum().sort_values(ascending = False)
```

8. Headed goals to goals

```python
def headgoalstogoals (row):
    row["Headed goals"] = (int(row["Headed goals"]) / int(row["Goals"]))
    return row

euro.apply(headgoalstogoals, axis='columns').groupby("Team")["Headed goals"].sum()
```

9. Total number of fouls

```python
def totalfouls (row):
    row["Fouls Won"] = (int(row["Fouls Won"]) + int(row["Fouls Conceded"]))
    return row

euro.apply(totalfouls, axis='columns').groupby("Team")["Fouls Won"].sum()
```

```python
10. How many completed Passes occured Offsides
```

```python
def peroff (row):
    row["Offsides"] = (int(row["Offsides"]) / int(row["Passes completed"]))
    return row

euro.apply(peroff, axis='columns').groupby("Team")["Offsides"].sum()
```
